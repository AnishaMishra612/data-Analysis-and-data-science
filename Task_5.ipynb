```python
# Importing required libraries.
import pandas as pd
import numpy as np
import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation
%matplotlib inline 
sns.set(color_codes=True)
```

```python
data=pd.read_csv('globalterrorism.csv')
data.head(10)
```

### Exploring and cleaning the data

```python
data.columns
```

```python
data.shape
```

```python
dframe= data[['iyear','imonth','extended','iday','country','country_txt','provstate','region','region_txt','city','latitude','longitude',
                 'success','gname','attacktype1_txt','targtype1_txt','nkill','nwound','targtype1','weaptype1_txt']]
dframe.head()
```

```python
dframe.shape
```

```python
dframe.dtypes
```

```python
# Rows containing duplicate data
duplicate_rows_df = dframe[dframe.duplicated()]
print('number of duplicate rows: ', duplicate_rows_df.shape)
```

```python
# Used to count the number of rows before removing the data
dframe.count() 
```

```python
dframe = dframe.drop_duplicates()
dframe.head(5)
```

```python
# Counting the number of rows after removing duplicates.
dframe.count()
```

```python
dframe.shape
```

```python
# Finding the null values.
dframe.isnull().sum()
```

```python
# Dropping the missing values.
dframe = dframe.dropna() 
dframe.count()
```

```python
# Finding the null values.
dframe.isnull().sum()
```

```python
dframe.describe()
```

## Starting the Exploratory data analysis

```python
y_count=dframe.groupby('iyear').sum()
display(y_count)
```

```python
dframe['iyear'].value_counts().count()
```

```python
years_count = data['iyear'].value_counts().sort_index()

display(years_count)
```

### Number of attacks by Years

```python
year = dframe['iyear'].unique()
plt.figure(figsize = (15,10))
sns.barplot(x = year,
           y = years_count)
plt.xticks(rotation = 50)
plt.xlabel('Year of attacks',fontsize=20)
plt.ylabel('Number of Attacks',fontsize=20)
plt.title('Attacks In Years',fontsize=30)
plt.show()
```

### Number of kills and wounds by each year

```python
fig=plt.figure(figsize=(30,15))
ax1=fig.add_subplot(121)
dframe.groupby("iyear")[["nkill","nwound"]].sum().sort_values(by = "iyear", ascending = False).plot(kind = "bar", color = ["black","green"],ax=ax1)
ax1.set_title("Kills and wound for each year")
```

#### Observation : 
The maximum wounded people are in the year 2015
The maximum killed people are in the year 2016



### Attacks by each region

```python
reg = dframe['region_txt'].unique()
reg_count = data['region_txt'].value_counts().sort_index()
plt.figure(figsize = (15,10))
sns.barplot(x = reg,
           y = reg_count)
plt.xticks(rotation = 50)
plt.xlabel('Region of attacks',fontsize=20)
plt.ylabel('Number of Attacks',fontsize=20)
plt.title('Attacks by each region',fontsize=30)
plt.show()
```
